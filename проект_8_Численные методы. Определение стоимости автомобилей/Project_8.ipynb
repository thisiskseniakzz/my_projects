### Визуализируем данные

```python
def plot_histograms(df, columns, color='skyblue', figsize=(20, 6), bins='auto'):
    for col in columns:
        plt.figure(figsize=figsize)
        
        # Обработка числовых столбцов
        if df[col].dtype in ['int64', 'float64']:
            # Гистограмма с KDE
            sns.histplot(data=df, x=col, color=color, bins=bins, kde=True)
            
            # Добавляем линии среднего и медианы
            mean_val = df[col].mean()
            median_val = df[col].median()
            
            plt.axvline(mean_val, color='red', linestyle='--', 
                        linewidth=2, label=f'Среднее: {mean_val:.2f}')
            plt.axvline(median_val, color='green', linestyle='-', 
                        linewidth=2, label=f'Медиана: {median_val:.2f}')
            
            # Статистика в заголовке
            stats = (f"min: {df[col].min():.0f} | max: {df[col].max():.0f} | "
                    f"std: {df[col].std():.2f} | пропуски: {df[col].isna().sum()}")
            plt.title(f'Распределение {col}\n{stats}', fontsize=14, pad=20)
            

        else:
            value_counts = df[col].value_counts()
            sns.barplot(x=value_counts.index, y=value_counts.values, color=color)
            

            for i, v in enumerate(value_counts.values):
                plt.text(i, v + 0.02*max(value_counts), str(v), 
                        ha='center', fontsize=10)
            
            plt.title(f'Значения{col}\n(Всего уникальных: {len(df[col].unique())})', 
                    fontsize=14, pad=20)
            plt.xticks(rotation=45, ha='right')
        
        plt.xlabel(col, fontsize=12)
        plt.ylabel('Количество', fontsize=12)
        plt.grid(axis='y', alpha=0.3)
        if df[col].dtype in ['int64', 'float64']:
            plt.legend()
        plt.tight_layout()
        plt.show()


columns_to_plot = [
    'price',               
    'vehicle_type',
    'gearbox',            
    'fuel_type', 
    'brand', 
    'repaired',
    'kilometer',
    'power',
    'registration_year'
]

# Вызываем функцию для визуализации
plot_histograms(df, columns_to_plot, color='skyblue')
```

## Обучение моделей

### Тренировочная выборка


```python
RANDOM_STATE = 42


# Разделение данных
X = df.drop('price', axis=1)
X = df.drop('date_created', axis=1)
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_STATE)

# Определение типов признаков
num_columns = ['registration_year','power','kilometer','registration_month']

# Категориальные признаки для OneHotEncoding (OHE)
ohe_columns = ['vehicle_type', 'gearbox', 'fuel_type','repaired', 'model'] 

# Создание пайплайна предобработки
preprocessor = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), num_columns),
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore'))
    ]), ohe_columns)
])



print(X_train.shape[0])
print(X_train.shape[1]) 

print(X_test.shape[0])
```

#### Ridge

```python
# Полный пайплайн_1
pipeline_ridge = Pipeline([
    ('preprocessor', preprocessor),
    ('model', Ridge())
])

# Параметры для GridSearch
param_grid_ridge = [
    {
        'model': [Ridge(random_state=RANDOM_STATE)],
        'model__alpha': [0.1, 1.0, 10.0],
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler()]
    }]
    

    # Поиск лучшей модели
grid_search_ridge = GridSearchCV(
    pipeline_ridge,
    param_grid_ridge,
    cv=5,
    scoring='neg_root_mean_squared_error', 
    refit=True,  
    n_jobs=-1,
    verbose=2
)

```

```python
%%timeit
grid_search_ridge.fit(X_train, y_train)
```

```python
%%timeit

best_model_ridge = grid_search_ridge.best_estimator_

# Предсказания и оценка на тренировочной выборке
y_train_pred = best_model_ridge.predict(X_train)

print("=== ЛУЧШАЯ МОДЕЛЬ RIDGE ===")
print(f"Параметры: {grid_search_ridge.best_params_}")
print(f"RMSE (кросс-валидация): {-grid_search_ridge.best_score_:.2f}")
```

#### LGBM

```python
# Полный пайплайн_2
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LGBMRegressor())
])


# Параметры для GridSearch
,
    # {
        #'model': [LinearRegression()],
        #'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler()]
   # },
param_grid_lgbm = [
    {
        'model': [LGBMRegressor(random_state=RANDOM_STATE)],
        'model__n_estimators': [50, 100],
        'model__max_depth': [5, 7],
        'model__learning_rate': [0.1, 0.05]
    }    
]

# Поиск лучшей модели
grid_search_lgbm = GridSearchCV(
    pipeline,
    param_grid_lgbm,
    cv=3,
    scoring='neg_root_mean_squared_error', 
    refit=True,  
    n_jobs=-1,
    verbose=2
)
```

```python
%%timeit
grid_search_lgbm.fit(X_train, y_train)
```

```python
%%timeit
# Лучшая модель
best_model_lgbm = grid_search_lgbm.best_estimator_
y_train_pred = best_model_lgbm.predict(X_train)


print("\nРезультаты на ТРЕНИРОВОЧНОЙ выборке:")
print(f"Лучшие параметры: {grid_search_lgbm.best_params_}")
print(f"RMSE: {-grid_search_lgbm.best_score_:.2f}")

```

Результаты на ТРЕНИРОВОЧНОЙ выборке:
Лучшие параметры: {'model': LGBMRegressor(max_depth=7, random_state=42), 'model__learning_rate': 0.1, 'model__max_depth': 7, 'model__n_estimators': 100}
RMSE: 1890.97
1.29 s ± 54 ms per loop (mean ± std. dev. of 7 runs, 1 loop each)

Мы подобрали лучшую модель для тестовых данных
